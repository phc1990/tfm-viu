```python
from astroquery.esa.xmm_newton import XMMNewton
from astropy.io import fits
from astropy.wcs import WCS
import pandas as pd
import os
```

```python

om_csv_file='/Users/eracero/Downloads/data_om_observations_unique.csv'
userdir = '/Users/eracero/'
om_df = pd.read_csv(om_csv_file, dtype={'observation_oid': object})


observation_id = om_df.observation_oid
#observation_id = [x[:-1] for x in orig_obsnames]
#print(observation_id)


data = []
row = []
filters=['MLIMV','MLIMB','MLIMU','MLIMW1','MLIMM2','MLIMW2']  
for id in observation_id:
    XMMNewton.download_data(id,instname='OM',name='OBSMLI',level='PPS',extension='FTZ')
    om_file=userdir+id+".tar"
    hdul = fits.open(om_file)
    row.append(id)
    for filter in filters:
        try:
            row.append(hdul[1].header[filter])
            
        except KeyError:
            row.append('')
            continue
    try:
        os.remove(om_file)
    except:
        print("Error while deleting file ", om_file)
    
    print("row=",row)
    data.append(row)
    row = []
print(data)

df = pd.DataFrame(data, columns=['observation_id','MLIMV','MLIMB','MLIMU','MLIMW1','MLIMM2','MLIMW2'])

print(df)
df.to_csv('om_limiting_mags.csv', index=False)
```
